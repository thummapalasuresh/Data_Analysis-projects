## MULTIPLE FEATURE PLOTS
TEMPO, LOUDENESS, ACOUSTICNESS, DANCEABLITY, DURATION, ENERGY, INSTRUMENTALNESS, LIVENESS, SPEECHNINESS, VALENCE

```python
intrest= ["tempo", "loudness","acousticness","danceability","duration_ms","energy","instrumentalness","liveness","speechiness","valence"]
for feature in intrest:
  pos=df[df["target"]==1] [feature]
  neg=df[df["target"]==0] [feature]

  plt.figure(figsize=(12,7))

  sns.distplot(pos,bins=30,label="positive",color="green")
  sns.distplot(neg,bins=30,label="negative",color="blue")

  plt.legend(loc="upper right")
  plt.title(f"positive and negative histogram plot for {feature}")
  plt.show()
```
