```python
dummies= pd.get_dummies(df['Gender'])
dummies.head(3)
```

#Joining this dummies dataframe to original dataframe                     by using "concat"

```python
df= pd.concat([df,dummies], axis=1)
df.head()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Product line', y='Female', data=df)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Product line',y='Male', data=df,palette='hot')
plt.xticks(rotation='vertical',size=15)
plt.show()
```

```python
place_df= pd.DataFrame(df['City'].value_counts())
place_df
```

```python
sns.barplot(x=place_df.index, y=place_df['City'], palette='Blues')
plt.show()
```

```python
payment_df= pd.DataFrame(df['Payment'].value_counts())
payment_df
```

```python
sns.barplot(x=payment_df.index, y=payment_df['Payment'], palette='cool')
plt.show()
```
