#**What is the overall sales trend?**

---



```python
df['month_year']=df['order_date'].apply(lambda x:x.strftime ('%Y-%m'))
```

```python
 df_trends=df.groupby('month_year').sum()['sales'].reset_index()
 df_trends
```

#**VIsualization**

```python
plt.figure(figsize=(16,5))
plt.plot(df_trends['month_year'],df_trends['sales'], color='g')
plt.xticks(rotation='vertical', size=8)
plt.show()
```

#**Which are the Top 10 products by sales?**

```python
df.columns
```

```python
df_products= pd.DataFrame(df.groupby('product_name').sum()['sales'])
df_sort=df_products.sort_values(by=['sales'], ascending=False)
df_sort.head(10)
```

#**#WHICH ARE THE MOST SELLING PRODUCTS?**

```python
df.columns
```

```python
best = pd.DataFrame(df.groupby('product_name').sum()['quantity'])
Sort= best.sort_values(by=['quantity'], ascending=False)
Sort.head(10)
Sort[:10]
```

#**WHAT IS THE MOST PREFERRED SHIP MODE?**

```python
plt.figure(figsize= (10,6))
sns.countplot(x='ship_mode', data=df)
plt.show()
```

#**WHICH ARE THE MOST PROFITABLE CATEGORY AND SUB-CATEGORY?**


```python
best_profit=pd.DataFrame(df.groupby(['category','sub_category']).sum()['profit'])
best_profit
```

```python
best_profit.sort_values(by=(['profit','category']), ascending=False)
```
